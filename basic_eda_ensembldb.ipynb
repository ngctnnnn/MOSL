```python
import os

MASTER_FOLDER = '.'
ensembldb_file = os.path.join(MASTER_FOLDER, 'genomics_ensembl_database.csv')
```

```python
import pandas as pd
ensembl_db = pd.read_csv(ensembldb_file)
ensembl_db
```

```python
ensembl_db = ensembl_db.drop(['Unnamed: 0', 'X'], axis=1)
ensembl_db
```

```python
protein_length = []
gene_length = []

for idx in range(len(ensembl_db)):
    gene_seq = ensembl_db['gene_sequence'][idx]
    protein_seq = ensembl_db['protein_sequence'][idx]

    if isinstance(protein_seq, float):
        protein_length.append(None)
        gene_length.append(None)
    else:
        protein_length.append(len(protein_seq))
        gene_length.append(len(gene_seq))

ensembl_db['protein_sequence_length'] = protein_length
ensembl_db['gene_sequence_length'] = gene_length
```

## Protein check

## Begins with methionine ?

```python
# # Is protein start with methionine
check_methionine = []
for idx in range(len(ensembl_db)):
    protein_seq = ensembl_db['protein_sequence'][idx]
    if isinstance(protein_seq, float):
        check_methionine.append(None)
    else:
        if protein_seq[0] != 'M':
            check_methionine.append(False)
        else:
            check_methionine.append(True)

ensembl_db['start_with_methionine'] = check_methionine
```

# Gene sequence

## Consists of premature stop codon ?

```python
def is_premature_stop(gene_seq):
    for idx in range(len(gene_seq) - 3):
        if gene_seq[idx:idx+3] in stop_codon:
            return True, idx
    return False, None
```

```python
stop_codon = ['TAG', 'TAA', 'TGA']

is_premature_stop_status = []
stop_index = []

for idx in range(len(ensembl_db)):
    gene_seq = ensembl_db['gene_sequence'][idx]
    if isinstance(gene_seq, float):
        is_premature_stop_status.append(None)
        stop_index.append(None)
    else:
        is_premature_stop_status.append(is_premature_stop(gene_seq)[0])
        stop_index.append(is_premature_stop(gene_seq)[1])

ensembl_db['is_premature_stop'] = is_premature_stop_status
ensembl_db['gene_sequencestop_index'] = stop_index
```

```python
ensembl_db
```

## Stop with a stop codon

```python
stop_codon = ['TAG', 'TAA', 'TGA']

is_stop_with_stop_codon = []

for idx in range(len(ensembl_db)):
    gene_seq = ensembl_db['gene_sequence'][idx]
    if isinstance(gene_seq, float):
        is_stop_with_stop_codon.append(None)
    else:
        if gene_seq[len(gene_seq) - 3:] in stop_codon:
            is_stop_with_stop_codon.append(True)
        else:
            is_stop_with_stop_codon.append(False)

ensembl_db['stop_with_stop_codon'] = is_stop_with_stop_codon
```
